#RelPlot

Método de alto nível para visualização de relações estatísticas entre variáveis. O método relplot() já combina o scatter plot e line plots.

Muito útil para ver relações entre duas ou mais variáveis com o uso de parâmetros de semântica dos dados.

Visualizando a relação entre a variável valor total da conta e gorjeta.

```
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips)
```

Plotamos um gráfico de dispersão onde é possível observar que quanto maior o valor da refeição maior é o valor da gorjeta. Assim podemos dizer que temos uma relação positiva entre essas duas variáveis.

No parâmetro data informamos de qual dataframe esses dados são provenientes.

Visualizando a relação entre as variáveis total da conta e gorjeta e colorindo cada ponto com a variável smoker (fumante).

```
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

A diferença aqui é que a coloração dos pontos está variando conforme a coluna smoker, conseguimos colocar três informações em um mesmo gráfico.

Basta informar para o parâmetro hue a coluna que vai colorir os pontos.

Visualizando a relação entre as variáveis total da conta e gorjeta e colorindo cada ponto com a variável Sex.

```
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips)
```

Agora a coloração dos pontos vem da coluna sex, temos uma maior concentração do sexo feminino com gastos menores e gorjetas menores também.

Visualizando a relação entre as variáveis total da conta e gorjeta e colorindo cada ponto com a variável sexo e definindo um estilo também baseado na variável sexo.

```
sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", data=tips)
```

O marcador dos pontos de dados agora difere de um sexo para o outro, se tivéssemos 3 tipos de variável nesse coluna por exemplo, teríamos 3 tipos de marcadores diferentes.

Nesse caso a identificação dos sexos fica ainda mais fácil ao bater os olhos no gráfico.

É possível representar quatro variáveis alterando a tonalidade e estilo de cada ponto. Mas é preciso cautela, pois, a visualização pode ficar confusa.

```
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", data=tips)
```

Até que seria interessante plotar esse tipo de informação, mas na verdade o gráfico ficou muito poluído dificultando a extração da informação de imediato.

Esse foi um exemplo em que as configurações escolhidas surtiram um efeito negativo em vez de informativo.

Em casos de variáveis numéricas para definir a tonalidade a paleta de cores altera de forma sequencial.

```
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

A tonalidade dos pontos foi alterando de acordo com o valor da coluna size que tem 4 valores distintos, existem poucos pontos por exemplo para o size=6.

É possível customizar a paleta de cores sequenciais no link:

https://seaborn.pydata.org/generated/seaborn.cubehelix_palette.html#seaborn.cubehelix_palette
Usando a opção size para definir o tamanho de cada ponto baseado em alguma coluna da base de dados.

```
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
```

Além da customização das cores dos pontos, podemos mudar o seu tamanho, e também colocá-los em função dos valores da coluna size.

Não fica tão nítida a diferença entre os tamanhos, vamos fazer uma alteração para que esse gráfico fique mais explicativo.

```
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15,300), data=tips)
```

Definimos que o tamanho mínimo será 15 e o máximo 300, esses valores serão distribuídos conforme os valores da coluna size.

Agora conseguimos identificar com mais clareza os pontos maiores e menores tornando o gráfico auto explicativo.

#Scatterplot

Caso não queira usar o relplot() o método scatterplot() já atende perfeitamente quando queremos ver relações nos dados.
A única diferença é que o scatterplot plota gráfico de dispersão apenas, logo este não contém a função lineplot().
Visualizando relação entre as 4 variáveis total_bill,tipo,day,smoker.

```
sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips)
```

Esse é mais um exemplo de que mais de 3 variáveis é um mesmo gráfico não funciona, mesmo tendo uma legenda que nesse tipo de gráfico já é definida como padrão.

É bom que você conheça o método scatterplot(), porém o método relplot() como vimos até aqui é bem mais completo e customizável.

#Cores no Seaborn

A função que controla as cores no seaborn e a color_palette().
Essa função fornece uma interface para gerar cores no seaborn.
Usamos esse método internamente ao usar o parâmetro palette.
O método color_palette() recebe o nome de uma paleta de cores do seaborn ou um colormap do matplotlib .
O método set_palette() recebe os mesmos parâmetros e é usado para definir um estilo padrão para todos os plots.
As cores devem ser definidas levando em consideração o tipo de dado que queremos plotar. (qualitativos, sequenciais,diverging).

*Ferramenta Color Brewer 2.0 * http://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3

Essa ferramenta não é da seaborn mas vai agregar um conhecimento extra sobre as paletas de cores, mostrando qual tipo de paleta deve ser usada para cada objetivo.

##Paleta de **cores**

###Dados quantitativos

```
sns.palplot(sns.color_palette("Accent"))
```

###Dados divergentes

```
sns.palplot(sns.color_palette("PuOr"))
```

###Dados sequenciais

```
sns.palplot(sns.color_palette("Blues"))
```

Gráfico usando a paleta de cor sequencial “Blues”



```
sns.relplot(x="total_bill", y="tip", hue="size", palette="Blues", size="size", sizes=(12,250), data=tips)
```

Gráfico usando a paleta de cor qualitativa “Set2”.





```
sns.catplot(x="sex", kind="count", palette="Set2", data=tips)
```

Gráfico usando a paleta de cor divergente “PiYG”.

```
sns.heatmap(tips.corr(), annot=True, cmap="PiYG")
```

As paletas também são ótimas para matrizes de correlação, onde a tonalidade da coloração identifica o grau de correlação entre as variáveis.

###Definindo cores padrão para gráficos Searborn.

```
sns.set_palette("pastel")
#plot usando a paleta de cor "pastel"
sns.relplot(x="total_bill", y="tip", data=tips);
```

Instrução with para definir cor temporariamente.




```
with sns.color_palette("Accent"):
  sns.relplot(x="total_bill", y="tip", size="size", sizes=(12,250), data=tips);
```

#Plotando dados categóricos

Quando trabalhamos com variáveis categóricas temos que visualizar dados de formas diferentes.
O Seaborn fornece o método catplot() que já contém diversos tipos de gráficos embutidos.
Isso facilita, pois, você pode usar diferentes gráficos usando um mesmo método.

##Gráficos de barras ou também conhecidos como gráficos de colunas.

```
sns.catplot(x="sex", kind="count", palette="Set2", data=tips)
```

##Gráfico de barras horizontais usando a coluna day

```
sns.catplot(y="day", kind="count", palette="Set1", data=tips)
```

##Scatter plot com dados categóricos

```
sns.catplot(x="day", y="total_bill", palette="Set2", data=tips)
```

O parâmetro swarm evita sobreposição de pontos.

```
sns.catplot(x="day", y="total_bill", kind='swarm', data=tips)
```

O parâmetro hue permite adicionarmos uma terceira variável a nossa visualização.

```
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", palette="Dark2", data=tips)
```

Podemos brincar com a tonalidade desses pontos, então temos um gráfico sem sobreposição de pontos e com coloração definida pela coluna sex.

*   O parâmetro order permite alterarmos a ordem padrão das categorias que estão sendo exibidas.
*   Isso é útil quando temos mais um gráfico na mesma figura e queremos manter as mesmas ordens.



```
sns.catplot(x="day", y="total_bill", kind='swarm', hue="sex", order=['Sat','Sun','Fri','Thur'], data=tips)
```

Invertendo a visualização para plots horizontais.

```
sns.catplot(x="total_bill", y="day", kind="swarm", data=tips)
```

#Gráficos com Regressão

Quando temos muitas variáveis quantitativas em nossos dados é interessante visualizar como estas se relacionam.
Podemos visualizar essas informações com linhas de regressão.
Com modelos de regressão simples é possível checar se existe alguma correlação entre algumas variáveis.

```
%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="sex", palette="Pastel1", data=tips)
```

Usando o parâmetro col para segregar os gráficos pelo valor da coluna categórica time

```
%matplotlib inline
sns.lmplot(x="total_bill",y="tip", hue="smoker", col="time", palette="Set2", data=tips)
```

Parâmetros scatter_kws e line_kws.

```
%matplotlib inline
sns.lmplot(x="total_bill",
                  y="tip", 
                 hue="smoker", 
                 col="time", 
                 palette="Set1",
                 data=tips, 
                 scatter_kws={'s': 90, 'alpha': 0.5},
                 line_kws={'lw': 4})
```

O mesmo gráfico com uma coloração diferente e dando uma enfase maior nas linhas de regressão, melhorando a visualização que tínhamos antes.

Segregando gráficos pelo parâmetro col e pelo parâmetro row .

```
%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="smoker",col="time", row="sex", palette="Set2", data=tips)
```

#Visualizando Outliers

##Boxplot de Dias da semana por total de conta.

```
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

Para cada dia da semana conseguimos ver o limite inferior e superior, os quartis, a mediana, como os dados estão distribuídos e os possíveis outliers (pontos pretos) que são valores que fogem da margem limite , valores atípicos no dia.

##Gráfico de Boxplot dos dias por total de conta e se a pessoa é fumante ou não.

```
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
```

##Tipo de Boxplot com foco maior na forma dos dados do que nos outliers.
Esse tipo de gráfico é interessante quando temos grandes datasets.

```
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips)
```

O Método stripplot permite plotar a distribuição dos dados.
Podemos combinar os dois gráficos para ter mais informação.

```
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips)
sns.stripplot(x="day", y="total_bill", data=tips, color="blue")
```

Os pontos mostram como está a distribuição dos dados dentro de cada boxplot, para uma massa maior de dados essa combinação não é indicada, pois deixa a visualização muito poluída.

##O método catplot() permite usarmos facetgrid assim podemos combinar gráficos em uma única figura.

```
%matplotlib inline
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=tips, kind="boxen",height=4)
```

#Gráficos de Correlação

Útil para visualizar se existem correlações positivas ou negativas entre colunas
Método corr() do pandas possibilita calcular correlação por tipo spearman ou pearson

```
correlacoes = tips.corr()
correlacoes
```

```
ax = sns.heatmap(correlacoes)
```

Quanto mais claro a cor maior a correlação e quanto mais escura menor é a correlação.

##Exibe os valores de correlação.

```
ax = sns.heatmap(correlacoes, annot=True)
```
